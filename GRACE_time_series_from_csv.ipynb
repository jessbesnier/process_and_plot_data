```python
#TWS with seasonality removed (anomalies)
lake = 'tres_iramos'
ds = pd.read_csv('E:/internship 2022/data/GRACE/csv_grace/no_seasonal/'+lake+'.csv')

ds['datetime']=pd.to_datetime(ds['date'])
df1 = ds.resample('M', on= 'datetime').mean()
```
